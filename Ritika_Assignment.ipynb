```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('us-marriages-divorces-1867-2014.csv')
data1
```

```python
data1['marriages_percapita'] = data1['Marriages_per_1000']/1000
data1['divorces_percapita'] = data1['Divorces_per_1000']/1000
```

```python
data1.plot(x="Year", y=["marriages_percapita", "divorces_percapita"])
plt.xlabel('Year')
plt.ylabel('Marriages/ Divorces per capita')
plt.xlim(1867, 2014)
```

```python
data_mod = data1[(data1['Year'] == 1900) | (data1['Year'] == 1950) |(data1['Year'] == 2000)]
```

```python

data_mod.plot(x= 'Year', y=["Marriages", "Divorces"], kind="bar")
```
